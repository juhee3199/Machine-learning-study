```python
soup
```

```python
title = soup.find_all(class_ = "sh_blog_title")
print(title)
```

```python
for i in title:
    print(i.attrs['title'])
    print(i.attrs['href'])
```

# Selenium 실습

```python
import requests
from bs4 import BeautifulSoup
import time

from selenium import webdriver

driver=webdriver.Chrome("./chromedriver.exe")
```

```python
driver.get('https://everytime.kr/')
```

```python
# driver.find_element_by_css_selector

# element: 단수. 제일 첫번째꺼 하나.
# elements: 복수. 리스트 형태로 여러 개.
```

```python
# 로그인 버튼 선택자
log_button = driver.find_element_by_css_selector("a.button.login")

log_button.click()
```

```python
id_ =" "
password_ = " "
```

```python
ID = driver.find_elements_by_css_selector("input.text")[0]
ID.send_keys(id_)

PW = driver.find_elements_by_css_selector("input.text")[1]
PW.send_keys(password_)

login = driver.find_elements_by_css_selector("input.text")[2]
login.click()
```

```python
time.sleep(2)  # 컴퓨터에서 딜레이 되는 시간이 있기 때문에. 딜레이를 주기 위함.

# find_element (0), elements(x)
more_page = driver.find_element_by_css_selector("#container > div.rightside > div:nth-child(3) > div > h3 > a")

more_page.click()
```

```python
# 핫 게시물 정보들을 수집
# 리스트 형태로 저장

contents_list=[]
contents = driver.find_elements_by_css_selector("a>p.medium")

for i in range(20):
    contents_list.append(contents[i].text)
```
